Looking for who our wrestlers have defeated in their last battle for the reign游녬

```python
url = "https://wwe.fandom.com/es/wiki/Campeonato_de_la_WWE"
```

```python
html = requests.get(url)
html
```

```python
soup = BeautifulSoup(html.content, "html.parser")
```

```python
tags = soup.find_all("table")
```

```python
tags = soup.find_all("table", attrs = {"class":"tabinfo"})
tags[0]
```

```python
defeated = pd.read_html(tags[0].prettify())
defeated[0]
```

```python
type(defeated[0])
```

```python
#Create a new variable for the DataFrame
defeated = defeated[0]
```

```python
#Rename columns from Spanish to English
defeated = defeated.rename(columns={"Luchador:":"Name",
                                      "Reinado N춿:": "Reign",
                                      "Derrot칩 a:":"Defeated",
                                      "Fecha:":"Date",
                                      "Show o Evento:":"Event",
                                     })
```

* Remove titles rows:

```python
defeated = defeated[defeated['Date'] != defeated['Event']]
```

```python
defeated.reset_index(drop=True, inplace=True)
```

```python
#Removing last row as there is no clear information for the analysis
defeated = defeated.iloc[:-1 , :]
defeated
```

```python
#Cleaning date column
defeated['Date'] = defeated['Date'].str.replace('de', '')
defeated
```

* Importing our first database of wrestlers:

```python
wrestlers = pd.read_csv("/Users/lauurasarabia/Ironhack/projects/Project-2/data/WWE_Wrestlers.csv")
wrestlers
```

```python
#Checking wrestlers columns types for being able to merge both tables
wrestlers.dtypes
```

```python
#Changing types to string
wrestlers = wrestlers.astype("str")
wrestlers.dtypes
```

* Merging both tables for our analysis:

```python
battles = pd.merge(wrestlers, defeated, on=["Name", "Reign"], how = "left")
battles[:20]
```

```python
battles.dropna(subset=['Defeated'], inplace=True)
```

```python
battles.columns
```

```python
battles.drop(["Days", "Event_x", "Unnamed: 0", "Location", "Event_y", "Date_y" ], axis=1, inplace=True)
```

```python
battles.rename(columns={'Date_x':"Date"},inplace=True)
```

```python
battles.reset_index(drop=True, inplace=True)
```

```python
#Merge table
battles
```

```python
battles["Defeated"].unique()
```

* Removing all multiple values in "Defeated" column:

```python
battles.drop(battles.index[[17, 31, 43, 46, 50, 57, 68, 70, 73, 74, 75,
                           77, 78, 80, 85, 88, 89, 91, 97, 106, 115, 119, 120]], inplace=True)
```

```python
battles.reset_index(drop=True, inplace=True)
```

```python
def get_horoscope(row):
    """
    It takes horoscope values related to "Name" column, finds same name on "Defeated" column and applies same "Horoscope" value
    """
    if row['Defeated'] in battles['Name'].values:
        return battles[battles['Name'] == row['Defeated']]['Horoscope'].values[0]
    else:
        return None

battles['Defeated_Horoscope'] = battles.apply(get_horoscope, axis=1)
print(battles)
```

```python
#Removing "None" values 
battles = battles.mask(battles.eq('None')).dropna()
```

```python
#Removing "0" values
battles = battles.mask(battles.eq('0')).dropna()
```

```python
battles.reset_index(drop=True, inplace=True)
```

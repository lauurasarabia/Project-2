API:

```python
love_calculator_key = os.getenv("access_key")
host = os.getenv("host")
```

```python
load_dotenv()
```

```python
url = "https://love-calculator.p.rapidapi.com/getPercentage?"
```

```python
headers = {
"X-RapidAPI-Key": f"{love_calculator_key}",
"X-RapidAPI-Host": f"{host}"
}
```

```python
# Create a new column to store the result
loveapi['Love_Percentage'] = None
```

```python
for i, row in loveapi.iterrows():
    fname = row['Name'].split()[0]
    sname = row['Defeated'].split()[0]
    querystring = {"fname": fname, "sname": sname}

    response = requests.get(url, headers=headers, params=querystring)

    data = response.json()
    love_percentage = data.get('result')
    loveapi.at[i, 'Love_Percentage'] = love_percentage
```
